```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('countries.csv')
data
```

```python
afghanistan = data[data.country=='Afghanistan']
plt.plot(afghanistan.year,afghanistan.lifeExpectancy)
plt.plot(afghanistan.year,afghanistan.population)
plt.xlabel('year')
plt.ylabel('GDP')
plt.title('GDP variation of afghanistan')
plt.legend(['lifeExpectancy','Population'])
plt.show()
```
